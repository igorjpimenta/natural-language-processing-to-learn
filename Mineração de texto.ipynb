# Mineração de texto

## Criação de um corpus

#### Baixar pacotes da biblioteca nltk

```python
import matplotlib.pyplot as plt
import nltk

# dowload all packages in nltk.download()
```

#### Importar arquivos para criar o corpus

```python
from nltk.corpus import PlaintextCorpusReader

corpus = PlaintextCorpusReader('Mineração de texto/dados', '.*') # .* seleciona todos os arquivos da pasta

arquivos = corpus.fileids()

display(arquivos)
```

#### Visualizar stopwords

```python
from nltk.corpus import stopwords

stopwords = stopwords.words('english')

display(stopwords)
```

#### Incluir pontuação nas stopwords

```python
from string import punctuation

[stopwords.append(l) for l in punctuation]

display(stopwords)
```

#### Criar um modelo de nuvem de palavras

```python
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud

mapa_cores = ListedColormap(['orange', 'green', 'red', 'magenta'])

wordcloud = WordCloud(background_color = 'white',
                      colormap = mapa_cores,
                      stopwords = stopwords,
                      max_words = 100)
```

#### Gerar nuvem de palavras

```python
wordcloud.generate(corpus.raw())

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

## Matriz de termos frequentes

#### Remover stopwords

```python
words = [w for w in corpus.words() if w not in stopwords]

print('%d palavras removidas (%.1f%%).' % (len(corpus.words()) - len(words), (len(words) / len(corpus.words()) * 100)))
```

#### Gerar termos frequentes

```python
frequencia = nltk.FreqDist(words)

display(frequencia.most_common(20))
```
